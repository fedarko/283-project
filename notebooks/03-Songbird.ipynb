# Run Songbird on the data, and extract a log-ratio to be used for prediction

```python
import pandas as pd
import biom
from biom.util import biom_open
```

## Prepare the input data

```python
metadata = pd.read_csv("../data/final_metadata.tsv", sep="\t", index_col=0)
metadata
```

```python
# Convert the cancer matrix file to a BIOM table. (The "cancer_table_with_header.tsv" file
# is the exact same data as in the GSE... matrix -- I just added the "FeatureID" header
# in the first row so that BIOM can parse the file properly --
# see https://github.com/biocore/biom-format/issues/841.)
with open("output/cancer_table_with_header.tsv", "r") as cancer_tsv_file:
    tbl = biom.Table.from_tsv(cancer_tsv_file, None, None, lambda x: x)
    
# Output that (full) BIOM table to a file if desired.
# with biom_open("output/cancer_table.biom", "w") as cancer_tbl_file:
#     tbl.to_hdf5(cancer_tbl_file, "Conversion from TSV")
```

```python
# Subset the sample metadata to just cancer samples, and then -- further -- to just cancer samples
# set as "Train".
cancer_metadata = metadata.loc[metadata["Status"] != "N"]
cancer_training_sample_metadata = cancer_metadata.loc[cancer_metadata["TrainTest"] == "Train"]
cancer_training_sample_metadata.to_csv("output/cancer_training_sample_metadata.tsv", sep="\t")
cancer_training_sample_metadata
```

```python
# Filter the BIOM table to just the training cancer samples.
just_training_cancer_sample_tbl = tbl.filter(cancer_training_sample_metadata.index, inplace=False)
```

```python
preselected_ids = pd.read_csv("../data/preselectedEnsemblIDList.txt", index_col=0, header=None)
preselected_ids
```

```python
# Also, filter the table to just the 750 preselected Ensembl IDs. We could of course try Songbird with or without
# first filtering to these IDs, but using a subset makes sense for a first pass (and since I am doing this on
# my beleaguered laptop, not a HPC).
just_training_cancer_sample_tbl_750ids = tbl.filter(preselected_ids.index, axis="observation", inplace=False)

# We'll use this one as the input for Songbird.
with biom_open("output/subset_cancer_train_table.biom", "w") as cancer_train_tbl_file:
    just_training_cancer_sample_tbl_750ids.to_hdf5(cancer_train_tbl_file, "Filtering to just training cancer samples and 750 Ensembl IDs")
```

```python
!qiime tools import \
    --input-path output/subset_cancer_train_table.biom \
    --output-path output/subset_cancer_train_table.qza \
    --type "FeatureTable[Frequency]"
```

## Run Songbird, on just the cancer training samples

We are deliberately *not* using the test data here. Our aim is to find a predictive log-ratio of genes from the training data that can be used as a classifier on the test data.

### Run a null model, for comparison

See the docs at https://github.com/biocore/songbird for details as to why this is useful.

```python
!qiime songbird multinomial \
    --i-table output/subset_cancer_train_table.qza \
    --m-metadata-file output/cancer_training_sample_metadata.tsv \
    --p-formula "1" \
    --p-epochs 5000 \
    --p-batch-size 10 \
    --p-num-random-test-examples 17 \
    --p-random-seed 333 \
    --p-summary-interval 0.01 \
    --o-differentials output/songbird-results/null-diffs.qza \
    --o-regression-stats output/songbird-results/null-regression-stats.qza \
    --o-regression-biplot output/songbird-results/null-regression-biplot.qza \
    --verbose
```

### Run a model using recurrent vs. nonrecurrent as the category to attempt to distinguish

```python
!qiime songbird multinomial \
    --i-table output/subset_cancer_train_table.qza \
    --m-metadata-file output/cancer_training_sample_metadata.tsv \
    --p-formula "C(Status, Treatment('C-N'))" \
    --p-epochs 5000 \
    --p-batch-size 10 \
    --p-num-random-test-examples 17 \
    --p-random-seed 333 \
    --p-summary-interval 0.01 \
    --o-differentials output/songbird-results/status-diffs.qza \
    --o-regression-stats output/songbird-results/status-regression-stats.qza \
    --o-regression-biplot output/songbird-results/status-regression-biplot.qza \
    --verbose
```
